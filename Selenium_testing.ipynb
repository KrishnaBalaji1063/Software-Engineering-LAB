```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Set up WebDriver (using headless Chrome to run in Colab)
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Run in headless mode (without opening a browser window)
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Create WebDriver instance
driver = webdriver.Chrome(options=options)

# Load the HTML file in the browser
driver.get("file:///content/quadratic_solver.html")

# Test 1: Hardcoded input
driver.find_element(By.ID, 'a').send_keys("1")
driver.find_element(By.ID, 'b').send_keys("-3")
driver.find_element(By.ID, 'c').send_keys("2")
driver.find_element(By.XPATH, "//input[@type='submit']").click()
time.sleep(2)
result = driver.find_element(By.ID, "result").text
print("Test 1 (Hardcoded input) result:", result)

# Test 2: Keyboard input (simulated)
driver.find_element(By.ID, 'a').clear()
driver.find_element(By.ID, 'b').clear()
driver.find_element(By.ID, 'c').clear()
driver.find_element(By.ID, 'a').send_keys("2")
driver.find_element(By.ID, 'b').send_keys("5")
driver.find_element(By.ID, 'c').send_keys("-3")
driver.find_element(By.XPATH, "//input[@type='submit']").click()
time.sleep(2)
result = driver.find_element(By.ID, "result").text
print("Test 2 (Keyboard input) result:", result)

# Test 3: Read from a file (example: coefficients.txt)
# Assuming you have a file named 'coefficients.txt' with content like:
# 1 -3 2
# 2 5 -3
# You need to simulate reading this file
with open("/content/coefficients.txt", "r") as f:
    lines = f.readlines()

for line in lines:
    a, b, c = map(float, line.split())
    # Clear previous inputs
    driver.find_element(By.ID, 'a').clear()
    driver.find_element(By.ID, 'b').clear()
    driver.find_element(By.ID, 'c').clear()

    # Input values from file
    driver.find_element(By.ID, 'a').send_keys(str(a))
    driver.find_element(By.ID, 'b').send_keys(str(b))
    driver.find_element(By.ID, 'c').send_keys(str(c))

    driver.find_element(By.XPATH, "//input[@type='submit']").click()
    time.sleep(2)
    result = driver.find_element(By.ID, "result").text
    print(f"Test 3 (File input) result for a={a}, b={b}, c={c}: {result}")

# Test 4: Single set of input (same as Test 2, but as a single input)
driver.find_element(By.ID, 'a').clear()
driver.find_element(By.ID, 'b').clear()
driver.find_element(By.ID, 'c').clear()
driver.find_element(By.ID, 'a').send_keys("3")
driver.find_element(By.ID, 'b').send_keys("4")
driver.find_element(By.ID, 'c').send_keys("-7")
driver.find_element(By.XPATH, "//input[@type='submit']").click()
time.sleep(2)
result = driver.find_element(By.ID, "result").text
print("Test 4 (Single input set) result:", result)

# Test 5: Multiple sets of inputs
input_sets = [
    (1, -3, 2),
    (2, 5, -3),
    (3, 4, -7)
]

for i, (a, b, c) in enumerate(input_sets, start=1):
    # Clear previous inputs
    driver.find_element(By.ID, 'a').clear()
    driver.find_element(By.ID, 'b').clear()
    driver.find_element(By.ID, 'c').clear()

    # Input values
    driver.find_element(By.ID, 'a').send_keys(str(a))
    driver.find_element(By.ID, 'b').send_keys(str(b))
    driver.find_element(By.ID, 'c').send_keys(str(c))

    driver.find_element(By.XPATH, "//input[@type='submit']").click()
    time.sleep(2)
    result = driver.find_element(By.ID, "result").text
    print(f"Test 5 (Multiple input set {i}) result:", result)

# Close the browser after tests
driver.quit()
```
